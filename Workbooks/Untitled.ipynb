```python
import datetime

# Extract necessary columns from the DataFrame
call_option_data = df[df['symbol'].str.endswith('CE')].copy()
call_option_data['time_to_expiration'] = None

# Calculate time to expiration for each call option
for index, row in call_option_data.iterrows():
    symbol = row['symbol']
    expiration_date_str = symbol[9:16]
    expiration_date = datetime.datetime.strptime(expiration_date_str, '%d%b%y').date()

    current_date_str = row['date']
    current_date = datetime.datetime.strptime(current_date_str, '%Y-%m-%d').date()

    time_to_expiration = (expiration_date - current_date).days
    call_option_data.loc[index, 'time_to_expiration'] = time_to_expiration

call_option_data[['symbol', 'time_to_expiration']].head()
```

```python
import py_vollib.black_scholes.implied_volatility as iv
import datetime
import numpy as np

call_option_data = df[df['symbol'].str.endswith('CE')].copy()
call_option_data['implied_volatility'] = 0.2

for index, row in call_option_data.iterrows():
    option_price = float(row['close'])
    stock_price = float(underlying_price)
    strike_price = float(row['symbol'].split('CE')[0][-5:])
    expiration_date_str = row['symbol'][9:16]
    expiration_date = datetime.datetime.strptime(expiration_date_str, '%d%b%y').date()
    current_date = datetime.datetime.strptime(row['date'], '%Y-%m-%d').date()
    time_to_expiration = (expiration_date - current_date).days / 365.0
    risk_free_rate = 0.065  
    initial_guess = 0.2  # Initial estimate of implied volatility

    try:
        implied_volatility = iv.implied_volatility(option_price, stock_price, strike_price, time_to_expiration, risk_free_rate, 'c')
        if np.isnan(implied_volatility):
            continue
        call_option_data.loc[index, 'implied_volatility'] = implied_volatility
    except:
        continue

# Print the resulting DataFrame with implied volatility values
print(call_option_data[['symbol', 'implied_volatility']])
```

```python
call_option_data
```

```python
import py_vollib.black_scholes.greeks.analytical as greeks
from datetime import datetime

# Calculate delta for each call option
for index, row in call_option_data.iterrows():
    option_price = row['close']
    stock_price = float(row['close'])
    strike_price = float(row['symbol'].split('CE')[0][-5:])
    expiration_date_str = row['symbol'][9:16]
    expiration_date = datetime.strptime(expiration_date_str, '%d%b%y').date()
    current_date = datetime.strptime(row['date'], '%Y-%m-%d').date()
    time_to_expiration = (expiration_date - current_date).days / 365.0
    risk_free_rate = 0.05
    volatility = row['implied_volatility']
    
#     print(underlying_price, stock_price, strike_price, expiration_date, current_date, time_to_expiration)
    delta = greeks.delta('c', underlying_price, strike_price, time_to_expiration, risk_free_rate, volatility)
    delta_rounded = round(delta, 3)
    call_option_data.loc[index, 'delta'] = delta_rounded

print(call_option_data[['symbol', 'delta']])
```

```python
call_option_data['delta_diff'] = abs(call_option_data['delta']-0.3)
```
